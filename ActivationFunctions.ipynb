## Motivation

In general, a neural network can be described via tensors and operators acting on these tensors. The simplest operator $\hat{O}$ acting on a general tensor $T$, that maps it into an *output* vector, can be simply a linear transformation $\Lambda$ with and/or a translation $b$, i.e., the $i$th component of the resulting output vector is obtained through:
  
$$
\begin{align} 
\left( \hat{O} \, T \right)^i= {\Lambda^{i}}_{n~m~k~\cdots} \, T^{n~m~k~\cdots} + b^i
\end{align}
$$



where we used the Einstein sum convention.
